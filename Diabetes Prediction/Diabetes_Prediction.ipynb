### **Diabetes Prediction using KNN**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import GridSearchCV, train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix, classification_report
import warnings
warnings.filterwarnings('ignore')
```

```python
# reading the value from the csv file.
df = pd.read_csv('diabetes.csv')
df.head()
```

We can see the range of the data are very uneven, we have to scale the data, as it improves the
performance of the model. 
##### **Scaling:** There are many methods available to scale the data. Some of the widely used methods are â†’ > 
> ***Standard Scaler:*** It makes the spread of the data 1, and mean 0, Its is not suitable if the distribution of the data is not gaussian or normal.

> ***Min Max Scaler:*** It shrinks the spread of the data between 0 and 1(incase of only positive values) or b/w -1 to 1 (incase of integer values)

> ***Robust Scaler:*** Its use case is same as MinMaxScaler() but it is not sensitive to outliers as the former one is.

```python
# Split the data into target and features:
X = df.drop('Outcome', axis = 1)
y = df.Outcome
```

```python
# Make a copy of the original feature DataFrame, It'll come in handy.
X_scaled = X.copy()
```

```python
# Take a look at the distribution of the data.
X.plot(kind = 'kde', subplots = False, figsize = (6, 6))
```

```python
# Initiate standard scaler
sc = StandardScaler()
X_scaled = sc.fit_transform(X_scaled)

# As fit_transform gives numpy array, to plot easily with pandas I've made the data into a dataframe.
X_scaled = pd.DataFrame(X_scaled)
X_scaled.columns = ['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness',
                    'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age']
```

```python
# Plot after scaling the data.
X_scaled.plot(kind = 'kde', subplots = False, figsize = (6, 6))
plt.figure(dpi = 600)
```

```python
# Splitting the data into training and testing data:
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, random_state = 1, stratify = y)
```

```python
# Instantiating a KNN Classifier model. [I like to use knc rather than knn for Classifier]
knc = KNeighborsClassifier()
```

To estimate the right number of n_neighbors weâ€™ll use a technique called ***Grid-SearchCV***.

```python
param_grid = {'n_neighbors': range(1, 50)}
grid = GridSearchCV(knc, param_grid, cv = 5)
grid.fit(X_train, y_train)
```

```python
# Printing the best score and best hyperparameter (here, n_neighbors)
print('Best Parameter: %s' % grid.best_params_)
print('Best Score: %s' % grid.best_score_)
```

```python
grid_prediction = grid.predict(X_test)
grid_prediction
```

```python
# We can construct a confusion matrix to visualize the correctly and incorrectly label scores:
conf_matrix = confusion_matrix(y_test, grid_prediction)
sns.heatmap(pd.DataFrame(conf_matrix), annot = True, cmap ='YlGnBu', fmt = 'g')
plt.title('Confusion matrix')
plt.ylabel('Actual label')
plt.xlabel('Predicted label')
plt.show()
```

```python
# We can get a detailed Classification Score of the model:
clf_report = classification_report(y_test, grid_prediction)
print(clf_report)
```

```python
all_scores = grid.cv_results_['mean_test_score']
plt.plot(range(1, 50), all_scores, 'ro-')
plt.xticks(range(1, 50))
plt.rcParams['figure.figsize'] = [25, 6]
plt.show()
```
